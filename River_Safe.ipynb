```python
df['ph']=df['ph'].fillna(df['ph'].mean())
df['Sulfate']=df['Sulfate'].fillna(df['Sulfate'].mean())
df['Trihalomethanes']=df['Trihalomethanes'].fillna(df['Trihalomethanes'].mean())
```

```python
df.isnull().sum()
```

```python
sns.heatmap(df.isnull())
```

```python
x=df.drop('Potability',axis=1)
y=df['Potability']
```

```python
x.shape,y.shape
```

```python
scaler=StandardScaler()
x=scaler.fit_transform(x)
```

```python
x
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train.shape,x_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
#creating object
model_lr=LogisticRegression()
```

```python
#training
model_lr.fit(x_train,y_train)
```

```python
# predictions
pred_lr=model_lr.predict(x_test)
```

```python
pred_lr
```

```python
#accuracy score
accuracy_score_lr=accuracy_score(y_test,pred_lr)
accuracy_score_lr
```

```python
from sklearn.tree import DecisionTreeClassifier
#object creation
model_dt=DecisionTreeClassifier(max_depth=4)
```

```python
#training
model_dt.fit(x_train,y_train)
```

```python
#prediction
pred_dt=model_dt.predict(x_test)
```

```python
#accuracy
accuracy_score_dt=accuracy_score(y_test,pred_dt)
accuracy_score_dt
```

```python
#confusion matrix
cm2=confusion_matrix(y_test,pred_dt)
cm2 #367 are 0s predicted correctly,38 wrong !! 205 are 1s predicted correctly,35 wrong
```

```python
from sklearn.ensemble import RandomForestClassifier
model_rf=RandomForestClassifier()
```

```python
#training
model_rf.fit(x_train,y_train)
```

```python
#prediction
pred_rf=model_rf.predict(x_test)
```

```python
#accuracy
accuracy_score_rf=accuracy_score(y_test,pred_rf)
accuracy_score_rf*100
```

```python
#confusion matrix
cm3=confusion_matrix(y_test,pred_rf)
cm3
```

```python
from sklearn.neighbors import KNeighborsClassifier
#creating model
# by default - model_knn=KNeighborsClassifier()
```

```python
for i in range (4,15):
model_knn=KNeighborsClassifier(n_neighbors=i)
model_knn.fit(x_train,y_train)
pred_knn=model_knn.predict(x_test)
accuracy_score_knn=accuracy_score(y_test,pred_knn)
print(i,accuracy_score_knn)
```

```python
model_knn=KNeighborsClassifier(n_neighbors=10)
model_knn.fit(x_train,y_train)
pred_knn=model_knn.predict(x_test)
accuracy_score_knn=accuracy_score(y_test,pred_knn)
print(accuracy_score_knn)
```

```python
from sklearn.svm import SVC#support vector classifier
```

```python
model_svm=SVC(kernel='rbf') #kernel types=linear,rbf(better accuracy),polynomial
```

```python
# prediction
pred_svm=model_svm.predict(x_test)
```

```python
#accuracy
accuracy_score_svm=accuracy_score(y_test,pred_svm)
accuracy_score_svm*100
```
